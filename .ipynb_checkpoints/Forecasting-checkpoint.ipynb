## Packages

```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima.arima import auto_arima
from arch import arch_model
import yfinance
import warnings
warnings.filterwarnings("ignore")
sns.set()
```

## Loading the data

```python
raw_data = yfinance.download (tickers = "^GSPC ^FTSE ^N225 ^GDAXI", start = "1994-01-07", 
                              end = "2019-09-01", interval = "1d", group_by = 'ticker', auto_adjust = True, treads = True)
```

```python
df_comp = raw_data.copy()
```

```python
df_comp['spx'] = df_comp['^GSPC'].Close[:]
df_comp['dax'] = df_comp['^GDAXI'].Close[:]
df_comp['ftse'] = df_comp['^FTSE'].Close[:]
df_comp['nikkei'] = df_comp['^N225'].Close[:]
```

```python
df_comp = df_comp.iloc[1:]
del df_comp['^N225']
del df_comp['^GSPC']
del df_comp['^GDAXI']
del df_comp['^FTSE']
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

## Creating Returns

```python
df_comp['ret_spx'] = df_comp.spx.pct_change(1).mul(100)
df_comp['ret_ftse'] = df_comp.ftse.pct_change(1).mul(100)
df_comp['ret_dax'] = df_comp.dax.pct_change(1).mul(100)
df_comp['ret_nikkei'] = df_comp.nikkei.pct_change(1).mul(100)
```

```python
df_comp['norm_ret_spx'] = df_comp.ret_spx.div(df_comp.ret_spx[1])*100
df_comp['norm_ret_ftse'] = df_comp.ret_ftse.div(df_comp.ret_ftse[1])*100
df_comp['norm_ret_dax'] = df_comp.ret_dax.div(df_comp.ret_dax[1])*100
df_comp['norm_ret_nikkei'] = df_comp.ret_nikkei.div(df_comp.ret_nikkei[1])*100
```

## Splitting the Data

```python
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## Fitting a Model

```python
model_1 = ARIMA(df.ftse, order=(1,0,0))
result_1 = model_1.fit()
```

## Simple Forecasting

### ARMAX

```python
# it largely improves the prediction
# however, this is unrealistic, as you wouldn't know the future values of other indexes, so your model would predict very poorly
# ARMAX would be very good at analyzing data, but when predicting (forecasting), it would be very poor
end_date = '2015-01-01'
model_11x = ARIMA(df.ret_ftse[1:], exog=df[['ret_spx', 'ret_dax', 'ret_nikkei']][1:], order=(1,0,1))
result_11x = model_11x.fit(start_ar_lags=14)

df_pred = result_11x.predict(start = start_date, end=end_date,
                            exog=df_test[['ret_spx', 'ret_dax', 'ret_nikkei']][start_date:end_date])

df_pred[start_date:end_date].plot(figsize=(20,5), color='red')
df_test.ret_ftse[start_date:end_date].plot(color='blue')
plt.title('Prediction vs. Actual', size=20)
plt.show()
```

### Forecasting Volatility

```python
# since that it will take automatically the validation set, we should specify the "last_obs" so it only train the data before
model_garch = arch_model(df_comp['ret_ftse'][1:], p=1, q=1, vol='GARCH', dist='Normal', mean='constant')
res_garch = model_garch.fit(update_freq=5, last_obs=start_date)
```

```python
# The first one is called Horizon and it determines how many observations we want our model to predict for each date since we're making predictions for each day.
# The second argument is called align. It determines whether we match the value with the date.
# The prediction is made on or the one it's supposed to represent. We prefer the latter, we set it 'target' as opposed to 'origin'
garch_pred = res_garch.forecast(horizon=1, align='target')
```

```python
from datetime import timedelta
garch_pred.residual_variance[start_date:str(pd.to_datetime(start_date)+timedelta(days=6))]
```

```python
df_test.head(4)
```

```python
# the GARCH is good at predicting the huge spike on 2016-2017
# zorder is to tell the order when the graph presenting line
garch_pred.residual_variance[start_date:].plot(figsize=(20,5), color='red', zorder=2)
df_test.ret_ftse[start_date:].abs().plot(color='blue', zorder=1)
plt.title('Volatility Prediction', size=20)
plt.show()
```

```python
# predict 100 days later, and we want to display from the last day in the data set
future_pred = res_garch.forecast(horizon=100, align='target')
future_pred.residual_variance[-1:]
```

### Multvariate Regression Model

```python
from statsmodels.tsa.api import VAR
```

```python
df_mul = df[['ret_ftse', 'ret_spx', 'ret_dax', 'ret_nikkei']][1:]
```

```python
model_var = VAR(df_mul)
model_var.select_order(20)
result_var = model_var.fit(ic = 'aic')
```

```python
# there are 4 models here for the four returns, and the multivariate regression just compute 5 lags in the model
result_var.summary()
```

```python
lag_order = result_var.k_ar
```

```python
var_pred = result_var.forecast(df_mul.values[-lag_order:], len(df_test[start_date:end_date]))

df_var_pred = pd.DataFrame(data=var_pred, index=df_test[start_date:end_date].index,
                          columns = df_test.columns[4:8])

df_var_pred.ret_nikkei[start_date:end_date].plot(figsize=(20,10), color='red')
df_test.ret_nikkei[start_date:end_date].plot(color='blue')
plt.title("Real vs Prediction", size = 24)
plt.show()
```
